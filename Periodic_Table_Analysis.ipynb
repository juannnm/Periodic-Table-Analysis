#**Periodic Table Analysis**

## Introduction
In this notebook we will be seeing some relationships and general information about the periodic table and the Elements.
The idea is to see some known insight on the elements and their properties and behaviors.

First we will import some useful libraries to help us manipulate the data and make some graphics. And then call from an url the raw data as a csv file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/juannnm/Periodic-Table-Analysis/main/periodic_table.csv'
pt = pd.read_csv(url)
pt.head(10)
```

```python
pt.describe()
```

```python
pt.replace(np.nan, 0.0, inplace=True)
```

## Structure Analysis
Lets see some graphs about this relationships

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=pt, x=pt['AtomicMass'], y=pt['AtomicNumber'], hue=pt['AtomicMass'])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=pt, x=pt['NumberofNeutrons'], y=pt['NumberofProtons'])
plt.show()
```

We can see the clearly correlation between the number of Neutron and the Atomic Mass

## Types and Phases of Elements
We will see the percentage of types and phases of elements.

```python
metalPercentage = (pt['Metal'] == 'yes').mean() * 100
print(metalPercentage)
nonMetalPercentage = (pt['Nonmetal'] == 'yes').mean() * 100
print(nonMetalPercentage)
metalloidPercentage = (pt['Metalloid'] == 'yes').mean() * 100
print(metalloidPercentage)
typesPercentage = (metalPercentage, nonMetalPercentage, metalloidPercentage)
```

```python
def determineType(x):
  if x['Metal'] == 'yes':
    return 'Metal'
  elif x['Nonmetal'] == 'yes':
    return 'NonMetal'
  else:
    return 'Metalloid'

pt['Element Type'] = pt.apply(determineType, axis=1)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(x=typesPercentage, labels=('Metal','NonMetal','Metalloid'), autopct='%1.1f%%', colors=['#26E0F0', '#F0D30E', '#F03EA4'])
plt.title('Comparision between Metals, NonMetals and Metalloids')
plt.axis('equal')
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=pt, x='Element Type', y='Electronegativity')
plt.show()
```

We can see two things here:
  First we see the direct relation between the electronegativity and the type of element
  The second thing we see is the exceptions to this, which are the Noble Gases. This ones have a value of 0 electronegativiy but still they can be categorize in one of this three groups. But if you pay attention the number of points with 0 Electronegativity are more that the number of Noble Gases, thats because some Artificial Elements donÂ´t have a registered Electronegativity, so we donÂ´t have data about that.

```python

color_palette = sns.color_palette("viridis", n_colors=pt['Period'].nunique())
sns.scatterplot(data=pt, x='NumberofElectrons', y='Electronegativity', hue='Period',palette=color_palette, s=100)
plt.show
```

```python
plt.figure(figsize=(6,6))
pt.groupby(['Phase']).count()['Element'].plot(kind='bar')
plt.xticks(rotation=35)
plt.title('Phase by Atomic Number')
plt.xlabel('Phases')
plt.ylabel('Atomic Number Sum')
plt.show()
```

In his graph we see that the Elements tend naturally to form structures in a solid state, followed by gaseous structures. The interesting thing here is that the artificial Elements, which have short medium lifes have a mayority compared to gaseous and liquid structures.

```python
plt.figure(figsize=(8,6))
pt.groupby(["Type"]).count()["Element"].plot(kind="bar")
plt.xticks(rotation=90)
plt.title("Type of Elements")
plt.xlabel("Types")
plt.ylabel("Elements Sum")
plt.show()
```

In this graph we can se the distribution between the types of the Elements
